All order Ids should exist in fact sales table

```python
%sql
SELECT COUNT(*) 
FROM data_engineering_project.silver_orders s
LEFT JOIN data_engineering_project.gold_fact_sales f
  ON s.order_id = f.order_id
WHERE f.order_id IS NULL;
```

Checking All customer, product and date key exist in dimension

```python
%sql
SELECT COUNT(*)
FROM data_engineering_project.gold_fact_sales f
LEFT JOIN data_engineering_project.gold_dim_customer c ON f.customer_key = c.customer_key
WHERE c.customer_key IS NULL

```

```python
%sql
SELECT *
FROM data_engineering_project.gold_fact_sales f
LEFT JOIN data_engineering_project.gold_dim_product p
ON f.product_key = p.product_key
WHERE p.product_key IS NULL

```

```python
%sql
SELECT COUNT(*)
FROM data_engineering_project.gold_fact_sales f
LEFT JOIN data_engineering_project.gold_dim_date d ON f.order_date_key = d.date_key
WHERE d.date_key IS NULL
```
